```python
import os
import pandas as pd
from config import username, password
import sqlalchemy
from sqlalchemy import create_engine
```

```python
filepath = "Resources/StudentsPerformance.csv"
student_perform = pd.read_csv(filepath)
```

```python
student_perform.head()
```

```python
new_stud_perform = student_perform.rename(columns = {
    "race/ethnicity" : "ethnicity",
    "parental level of education" : "parent_edu",
    "test preparation course" : "test_prep",
    "math score": "math_score",
    "reading score": "reading_score",
    "writing score": "writing_score"
})
```

```python
new_stud_perform.head()
```

```python
# Connect To Postgres
connection_string = f"{username}:{password}@localhost:5432/student_performances"
engine = create_engine(f'postgresql://{connection_string}')
conn = engine.connect()
```

```python
# Insert DF Into Postgres
new_stud_perform.to_sql('student_scores', con = engine, if_exists = 'append', index=False)
```

```python
# Cheack Connection From Postgres to Pandas
data = pd.read_sql('SELECT * FROM student_scores', conn)
data.head(5)
```
